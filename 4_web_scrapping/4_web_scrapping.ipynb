# Обязательная часть
Вам необходимо написать функцию, которая будет основана на поиске по сайту habr.com. Функция в качестве параметра должна принимать список запросов для поиска (например, ['python', 'анализ данных']) и на основе материалов, попавших в результаты поиска по каждому запросу, возвращать датафрейм вида:  

<дата> - <заголовок> - <ссылка на материал>  
В рамках задания предполагается работа только с одной (первой) страницей результатов поисковой выдачи для каждого запроса. Материалы в датафрейме не должны дублироваться, если они попадали в результаты поиска для нескольких запросов из списка.


```python
base_url = 'https://habr.com'
def hapr_posts(list_query):
    ' функция для парсинга с хабра '
    # создадим словарь, куда будем складывать нужные данные
    df_dict = {'дата': [],
               'заголовок': [],
               'ссылка на материал': []}
    # ищем данные для каждого запроса
    for query in list_query:
        # создаем ссылку и получаем данные по ней
        query_url = base_url + '/ru/search/?q=' + query + '&target_type=posts&order=relevance'
        response = requests.get(query_url)
        # Парсим HTML при помощи Beautiful Soup
        soup = BeautifulSoup(response.text, 'html.parser').find('div', {'class': 'tm-articles-list'})
        # для каждой статьи извлекаем необходимые данные
        for row in soup.find_all('article', {'class': 'tm-articles-list__item'}):
            df_dict['дата'].append(row.time['title'].split(',')[0])
            df_dict['заголовок'].append(row.find('h2').text)
            df_dict['ссылка на материал'].append(base_url + row.find('h2').a.get('href'))
    # возвращаем датафрейм, удалив дубли
    return pd.DataFrame(df_dict).drop_duplicates()
```

```python
df = hapr_posts(['python', 'анализ данных'])
```
